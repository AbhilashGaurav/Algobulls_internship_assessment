### # Class with functionlaity request in the point 1

```python
import requests
import pandas as pd

class ScriptData:
    def __init__(self):
        self.api_key = "9H5W499KUM6FLCG1"
        self.data = {}
    
    def fetch_intraday_data(self, script):
        url = f"https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY&symbol={script}&interval=60min&apikey={self.api_key}"
        print(url)
        response = requests.get(url)
        self.data[script] = response.json()
    
    def convert_intraday_data(self, script):
        if script not in self.data:
            raise ValueError(f"Data for script '{script}' not found. Please fetch data using the 'fetch_intraday_data' method.")
        time_series = self.data[script]["Time Series (60min)"]
        data = []
        for timestamp, values in time_series.items():
            data.append([pd.Timestamp(timestamp), float(values["1. open"]), float(values["2. high"]), 
                         float(values["3. low"]), float(values["4. close"]), int(values["5. volume"])])
        self.data[script] = pd.DataFrame(data, columns=["timestamp", "open", "high", "low", "close", "volume"])
    
    def __getitem__(self, script):
        if script not in self.data:
            raise ValueError(f"Data for script '{script}' not found. Please fetch and convert data using the 'fetch_intraday_data' and 'convert_intraday_data' methods.")
        return self.data[script]
    
    def __setitem__(self, script, df):
        self.data[script] = df
    
    def __contains__(self, script):
        return script in self.data
```

**2.**Define a function called indicator1. It should take “df” and ‘timeperiod’ (integer) as inputs and give another pandas DataFrame as an output with two columns:

**a. timestamp:** Same as ‘timestamp’ column in ‘df’

**b. indicator:** Moving Average of the ‘close’ column in ‘df’. The number of
elements to be taken for a moving average is defined by ‘timeperiod’. 

For example, if ‘timeperiod’ is 5, then each row in this column will be an average
of total 5 previous values (including current value) of the ‘close’ column.

________________________________________________________________________________



### # Function declartion for indicator1

```python
def indicator1(df, timeperiod):
    df['indicator'] = df['close'].rolling(window=timeperiod).mean()
    return df[['timestamp', 'indicator']]
```

### # Class Strategy

```python
class Strategy:
    def __init__(self, script_name):
        self.script_name = script_name
        self.df = None
        self.df1 = None
        self.indicator_df = None
        self.signals = None
        self.script_data = ScriptData()
    # creating the scriptdata for close_data
    def fetch_data(self):
      self.script_data.fetch_intraday_data(self.script_name)
      self.script_data.convert_intraday_data(self.script_name)
      self.df = self.script_data[self.script_name]
      self.df.rename(columns={'close':'close_data'},inplace=True)
      # print(self.df)
      return self.df

    # creating indicator data on close of script data
    def indicator1(self,timeperiod=5):
      self.df=self.fetch_data()
      df=self.df
      df['indicator'] = df['close_data'].rolling(window=timeperiod).mean()
      return df[['timestamp', 'indicator']]
    def generate_signals(self, timeperiod):
        self.df1=self.indicator1(timeperiod)
        # print(self.df1)
        signals = []
        for i in range(1, len(self.df)):
            if self.df1['indicator'][i] > self.df['close_data'][i] and self.df1['indicator'][i-1] <= self.df['close_data'][i-1]:
                signals.append('BUY')
            elif self.df1['indicator'][i] < self.df['close_data'][i] and self.df1['indicator'][i-1] >= self.df['close_data'][i-1]:
                signals.append('SELL')
            else:
                signals.append('NO_SIGNAL')
        self.signals = pd.DataFrame({'timestamp': self.df['timestamp'][1:], 'signal': signals})
    
    def print_signals(self):
        # signals_df = self.signals[self.signals['signal'] != 'NO_SIGNAL']
        signals_df = self.signals
        print(signals_df )
```

### #Query 3 d

```python
#  d
strategy = Strategy('NVDA')
strategy.generate_signals(5)
strategy.print_signals()
```
